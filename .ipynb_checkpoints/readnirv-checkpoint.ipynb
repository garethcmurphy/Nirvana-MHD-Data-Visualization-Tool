```python
%matplotlib inline
from __future__ import print_function
import numpy as np
import h5py
import matplotlib.pyplot as plt

with h5py.File('usr790000.h5','r') as hf:
    print('List of arrays in this file: \n', hf.keys())
    dens = hf.get('density')
    den = np.transpose(np.array(dens))
    bh=hf.get('magnetic_field')   
    vh=hf.get('velocity')   
    b=bh.value
    v=vh.value
    vx=np.transpose(v[:,:,:,0])
    vy=np.transpose(v[:,:,:,1])
    vz=np.transpose(v[:,:,:,2])
    bx=np.transpose(b[:,:,:,0])
    by=np.transpose(b[:,:,:,1])
    bz=np.transpose(b[:,:,:,2])


de2d=np.log10(den[:,:,0])
vx2d= vx[:,:,0]
vy2d= vy[:,:,0]
vz2d= vz[:,:,0]
bx2d= bx[:,:,0]
by2d= by[:,:,0]
bz2d= bz[:,:,0]


    
```

```python

#plt.imshow(arr)
from matplotlib import font_manager 
font_manager.findfont('Times New Roman') 
plt.rc('font', family='serif') 
plt.rc('font', serif='Times New Roman') 

#f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
#f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')
f, ( ax1, ax2 ,ax3, ax4, ax5, ax6, ax7 ) = plt.subplots(1, 7,  sharey='row', figsize=(12,5))

ax1.imshow( de2d, origin='lower')
ax1.set_title(r'$\rho$')
start, end = ax1.get_xlim()
stepsize=100
start=0
end=300
ax1.xaxis.set_ticks(np.arange(start, end, stepsize))
ax2.imshow( vx2d)
ax2.set_title('Vx')
ax2.xaxis.set_ticks(np.arange(start, end, stepsize))

ax3.imshow( vy2d)
ax3.set_title('Vy')
ax3.xaxis.set_ticks(np.arange(start, end, stepsize))

ax4.imshow( vz2d)
ax4.set_title('Vz')
ax4.xaxis.set_ticks(np.arange(start, end, stepsize))

ax5.imshow( bx2d)
ax5.set_title('Bx')
ax5.xaxis.set_ticks(np.arange(start, end, stepsize))

ax6.imshow( by2d)
ax6.set_title('By')
ax6.xaxis.set_ticks(np.arange(start, end, stepsize))

ax7.imshow( bz2d)
ax7.set_title('Bz')
ax7.xaxis.set_ticks(np.arange(start, end, stepsize))


plt.savefig("test.pdf")
```
